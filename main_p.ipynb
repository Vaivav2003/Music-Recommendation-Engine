```python
dataset.info()
dataset.drop_duplicates(subset=['name'],inplace=True)
dataset.shape
```

```python
# removing name ,release date, id
dataset.drop(columns=['name','release_date','id'],inplace=True)
dataset.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
dataset['artists']=encoder.fit_transform(dataset['artists'])
dataset['artists']
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
columns_to_standardize = ['valence', 'acousticness', 'danceability', 'duration_ms',
                          'energy', 'instrumentalness', 'liveness', 'loudness',
                          'popularity', 'speechiness', 'tempo']

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the selected columns and transform them
dataset[columns_to_standardize] = scaler.fit_transform(dataset[columns_to_standardize])
dataset.describe()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score
model=KMeans(n_clusters=10,init='k-means++')
model.fit(dataset)
calinski_harabasz = calinski_harabasz_score(dataset, model.labels_)
davies_bouldin = davies_bouldin_score(dataset, model.labels_)
print(f"Calinski-Harabasz Index: {calinski_harabasz}")
print(f"Davies-Bouldin Index: {davies_bouldin}")
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_clusters':[i for i in range(2,100)]}
clst = GridSearchCV(model, parameters)
clst.fit(dataset)
```

```python
# Print the best parameters found by grid search
print("Best parameters:", clst.best_params_)

# Get the best estimator (KMeans model) found by grid search
best_model = clst.best_estimator_

# Fit the best model to the dataset
best_model.fit(dataset)

# Access cluster labels assigned by the best model
cluster_labels = best_model.labels_
```

```python
best_model.inertia_
```

```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
# Define a pipeline with PCA and KMeans
pipeline = Pipeline([
    ('pca', PCA()),       # Step 1: PCA
    ('kmeans', KMeans())  # Step 2: KMeans
])

# Define the parameter grid for grid search
parameters = {
    'pca__n_components': [i for i in range(2,16)],  # Number of principal components
    'kmeans__n_clusters': [ i for i in range(2,1000)]  # Number of clusters
}

# Create a GridSearchCV instance with the pipeline
grid_search = GridSearchCV(pipeline, parameters)
grid_search.fit(dataset)
```

```python
# Print the best parameters found by grid search
print("Best parameters:", grid_search.best_params_)

# Get the best estimator (KMeans model) found by grid search
best_model = grid_search.best_estimator_

# Fit the best model to the dataset
best_model.fit(dataset)

# Access cluster labels assigned by the best model
# cluster_labels = best_model.labels_
```
